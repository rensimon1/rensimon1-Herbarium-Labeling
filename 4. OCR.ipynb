```python
import os
from IPython.display import Image as IPImage

def traiter_images_ocr(dossier_etiquettes, output_dir, reader):
    """
    Traite les images d'un dossier spécifié pour la reconnaissance de texte et sauvegarde les résultats dans un dossier de sortie.

    Args:
        dossier_etiquettes (str): Le chemin du dossier contenant les images à traiter.
        output_dir (str): Le chemin du dossier où les fichiers de texte résultants doivent être sauvegardés.
        reader (EasyOCR.Reader): Une instance de Reader de EasyOCR configurée pour la reconnaissance de texte.

    Description:
        Cette fonction parcourt toutes les images dans le dossier spécifié, utilise EasyOCR pour extraire le texte de chaque image,
        et sauvegarde le texte dans un fichier .txt correspondant dans le dossier de sortie spécifié.
        Chaque fichier texte porte le même nom que l'image source mais avec l'extension .txt.
        La fonction affiche également chaque image après traitement pour vérification visuelle.
    """
    count = 1  # Initialiser le compteur d'images traitées
    os.makedirs(output_dir, exist_ok=True)  # Créer le dossier de sortie si nécessaire
    image_files = [f for f in os.listdir(dossier_etiquettes) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    for fichier in image_files:
        chemin_complet = os.path.join(dossier_etiquettes, fichier)
        print(f"Traitement de l'image {count}/{len(image_files)} : {fichier}")
        count += 1

        # Lire le texte de l'image avec EasyOCR
        result = reader.readtext(chemin_complet, detail=0, paragraph=True)
        print(result)

        # Sauvegarder le résultat du texte dans un fichier texte
        nom_fichier_sortie = os.path.splitext(fichier)[0] + ".txt"
        chemin_sortie = os.path.join(output_dir, nom_fichier_sortie)
        with open(chemin_sortie, 'w', encoding='utf-8') as f:
            f.write("\n".join(result))

        # Afficher l'image traitée pour vérification visuelle
        display(IPImage(filename=chemin_complet))
```

```python
import os
from PIL import Image
import pytesseract
from IPython.display import Image as IPImage

def traiter_images_tesseract(dossier_etiquettes, output_dir):
   """
    Traite les images d'un dossier spécifié pour la reconnaissance de texte avec Tesseract OCR et sauvegarde les résultats dans un dossier de sortie.

    Args:
        dossier_etiquettes (str): Le chemin du dossier contenant les images à traiter.
        output_dir (str): Le chemin du dossier où les fichiers de texte résultants doivent être sauvegardés.

    Description:
        Cette fonction parcourt toutes les images dans le dossier spécifié, utilise Tesseract OCR pour extraire le texte de chaque image,
        et sauvegarde le texte dans un fichier .txt correspondant dans le dossier de sortie spécifié.
        Chaque fichier texte porte le même nom que l'image source mais avec l'extension .txt.
        La fonction affiche également chaque image après traitement pour vérification visuelle.
        La fonction initialise un compteur pour suivre le nombre d'images traitées et affiche le progrès dans la console.
    """
    count = 0
    image_files = [f for f in os.listdir(dossier_etiquettes) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    os.makedirs(output_dir, exist_ok=True)
    for fichier in image_files:
        chemin_complet = os.path.join(dossier_etiquettes, fichier)
        print(f"Traitement de l'image {count + 1}/{len(image_files)} : {fichier}")
        count += 1

        # Lire le texte avec TesseractOCR
        image = Image.open(chemin_complet)
        result = pytesseract.image_to_string(image, lang='eng')
        print(result)

        # Sauvegarder le résultat dans un fichier texte
        nom_fichier_sortie = os.path.splitext(fichier)[0] + ".txt"
        chemin_sortie = os.path.join(output_dir, nom_fichier_sortie)
        with open(chemin_sortie, 'w', encoding='utf-8') as f:
            f.write(result)

        # Afficher l'image
        display(IPImage(filename=chemin_complet))

```
